```python
class NodoDoble:
    def __init__(self , value, siguiente = None, previo = None ):
                 self.data = value
                 self.next = siguiente
                 self.prev = previo 
```

```python
#ejercicio 1, este es el nodo doble
head = NodoDoble(10)
tail = head
# dos maneras de imprimir el nodo 
print( head.data )
print( tail.data )

# ejercicio 2 crear dos nodos imprimir el valor de 10 imprimir el 2do valor y hacer un tranversal
print("Ejercicio 2")
head = NodoDoble(10 , NodoDoble(20))
head.next.prev = head
tail = head.next
# doas maneras de imprimirlo
print("El primero :" , head.data)
print("El primero :" , tail.prev.data )

# ejercicio 3, imprimir el 2do valor 
print("ejercico 3")
print("El segundo :" , head.next.data)
print("El segundo :" , tail.data )

print("ehercicio 3 ")
head = NodoDoble(10, NodoDoble(20, NodoDoble(30, NodoDoble(40))))
# poner tail al ultimo
tail = head.next.next.next
head.next.prev = head
head.next.next.prev = head.next
tail.prev = head.next.next
print("El primero :" , head.data)
print("El primero :" , tail.prev.prev.prev.data )

print("impirimir el ultimo ejercicio 4 ")
print("El ultimo :" , tail.data)
print("El ultimo :" , head.next.next.next.data )

# transversal
print("Ejercicio 5, hacer el transversal")
curr_node = head
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.next
print("")

# reverse transversal
curr_node = tail
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.prev
print("")

print("agregar el 50 al final")
tail.next = NodoDoble( 50, None, tail)
curr_node = head
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.next
print("")

curr_node = tail.next
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.prev
print("")

```

```python
class DoubleLinkedList:
    def __init__( self ):
        self.head = None
        self.tail = None
        self.size = 0

    def get_size( self ):
        return self.size
    
    def is_empty( self ): # true si la lisat esta vacia
        return self.size == 0 # si es asi la lista esta vacia

    def append(self, value ):
        if self.is_empty():
            self.head = NodoDoble( value )
            self.tail = self.head

        else:
            self.tail.next = NodoDoble( value , None , self.tail )
            self.tail = self.tail.next
        self.size += 1


    def transversal(self):
        curr_node = self.head
        while curr_node != None :
            print(f"| {curr_node.data} |--->" , end="")
            curr_node = curr_node.next
        print("")

    def reverse_transversal(self):
        curr_node = self.tail
        while curr_node != None:
            print(f"| {curr_node.data} |--->" , end="")
            curr_node = curr_node.prev
        print("")


    def remove_from_tail( self , value ):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next

        if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev 

        self.size -= 1


    def remove_from_head( self , value  ):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next

        if curr_node.data == value:
            print("Encontrado",curr_node.data)
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
            print("", curr_node.prev.data)
        self.size -= 1

    def find_from_tail( self, value ):
         curr_node = self.tail
         while curr_node.data != value:
            curr_node = curr_node.prev

         if curr_node.data == value:
            print("Encontrado :",curr_node.data)
           
    def find_from_head( self, value ):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next

        if curr_node.data == value:
            print("Encontrado :",curr_node.data)
            
        
    
```
